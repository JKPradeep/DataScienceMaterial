### 1. Normalization (min-max Normalization)

In this appraoach we need to scale down the values of the features between 0 to 1

Xnorm  =  (X-  Xmin )/ (Xmax-  Xmin )

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()  # Create Object of MinMax Scaler
scaling.fit_transform(df4[['Alcohol','Malt']])
```

### 2. Standardizarion (Z-score Normalization)

Here all the features will be transformed in such a way that it will have the properties of a standard normal distribution with mean(μ)=0 and standard deviation(σ)=1
- z = (X — μ) / σ

```python
from sklearn.preprocessing import StandardScaler
scaling = StandardScaler()
scaling.fit_transform(df4[['Alcohol','Malt']])
```
