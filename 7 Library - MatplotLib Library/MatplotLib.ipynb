# MatplotLib
MatplotLib is a plotting library for the Python Programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+.

#### Some of the major pros of MatlabLib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High quality output in many formats
- Very Customizable in general

#### Scatter Plot

```python
# Plotting using matplotlib
# Scatter -  Plots in a scatter manner 
plt.scatter(x,y,c='r',marker='x',s=80)  # c-color; m-marker symbol; s-marker size
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Sample Graph')
plt.savefig('sampleGraph.png')     # To save Figure - Using Inbuilt Function / RightClick on Image 
plt.savefig('sampleGraph.jpeg')
```

#### Pie Charts

#### Multiple Axes

```python
fig = plt.figure()
ax = fig.add_axes([0.5,0.5,1,1]) # We can't see clearly MultipleAxes in Jupyter Notebook
# Refer : http://www.learningaboutelectronics.com/Articles/How-to-add-multiple-axes-to-a-figure-in-Python.php
cars['mpg'].plot()
index=np.arange(len(cars['C_name']))
plt.xticks(index,cars.C_name,rotation=90) # Total No. of Names given using Index Length 
ax.set_yticklabels(cars.mpg,rotation=60,fontsize='medium')
ax.set_title('Miles per gallon of cars in Cars')
ax.set_xlabel('carnames')
ax.set_ylabel('miles/gal')
ax.legend(loc='best')
ax.annotate('Honda Civic',xy=(18,30.4),arrowprops=dict(facecolor='blue'))
ax.annotate('Fiat X1-9',xy=(25,27.3),arrowprops=dict(facecolor='g'))
```

#### function

```python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.5)

plt.figure(figsize=(10,5))
plt.plot(t1, f(t1), color='red',marker='*')
plt.plot(t2, f(t2), color='blue',marker='o')
```
